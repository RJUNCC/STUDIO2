## Import dependencies
---

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## Get URL
---

```python
url = "https://webpages.charlotte.edu/mscipion/"
r = requests.get(url)
print(r)
```

## Create soup object
---

```python
# Create soup object
soup = BeautifulSoup(r.text, 'lxml')
```

## Find table tag and table headers
---

```python
table = soup.find("table")
# table
```

```python
# find the column headers
header = table.find_all('th')
header
```

## Put the headers into a list
---

```python
# put column titles into a list 
titles = []

for i in header:
    title = i.text
    titles.append(title)

titles
```

## Find all values in table row tag
---

```python
rows = table.find_all('tr')
# rows
```

```python
rows[1].find_all('td')[:3]
```

## Put it into a dataframe
---

```python
df = pd.DataFrame(columns=titles) # i put the df in here because everytime i ran the cell more than once and it kept increasing the length of df
for i in rows[1:]:
    data = i.find_all('td')
    row = [tr.text for tr in data]
    l = len(df)
    df.loc[l] = row

len(df)
```

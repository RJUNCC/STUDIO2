```python
# 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x, y = np.meshgrid(np.linspace(-5, 5, 10),
                   np.linspace(-5, 5, 10))

u = -x
v = -y

plt.quiver(x, y, u, v, color='g')
plt.title('Vector field')

plt.xlim(-7, 7)
plt.ylim(-7, 7)

plt.grid()
plt.show()
```

```python
from sympy.physics.vector import ReferenceFrame
from sympy.physics.vector import curl
import scipy as sp
import numdifftools as nd

R = ReferenceFrame('R')

F = R[1]**2 * R[2] * R.x - R[0]*R[1] * R.y + R[2]**2 * R.z

G = curl(F, R)  
```

```python
G  # 0i + y^2j + (-2yz-y)k = <0, y^2, -2yz-y>
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

x, y, z = np.meshgrid(np.arange(-1.0, 1.3, 0.2),
                      np.arange(-1.0, 1.3, 0.2),
                      np.arange(-1.0, 1.3, 0.8))

u = x**3 * y**4 * z**4
v = x**4 * y**3 * z**4
w = x**5 * y**8 * z**6

ax.quiver(x, y, z, u, v, w, color='g')
plt.title('Vector field')

# plt.xlim(-7, 7)
# plt.ylim(-7, 7)

plt.grid()
plt.show()
```

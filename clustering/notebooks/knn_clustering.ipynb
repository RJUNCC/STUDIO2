```python
from sklearn.neighbors import KNeighborsRegressor

import numpy as np
```

```python
X = np.array([
    [35, 30000],
    [45, 45000],
    [40, 50000],
    [35, 35000],
    [25, 32000],
    [40, 40000]
])
```

```python
KNN = KNeighborsRegressor(n_neighbors=3).fit(X[:, 0].reshape(-1, 1), X[:, 1])
```

```python
result = KNN.predict([
    [30]
])
```

```python
print(result)
```
